```python
# ================================
# PART 1 — DATA CLEANING
# ================================

# # Raw data given in the assignment
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

# Loop through each student
for student in raw_students:
    
    # 1. Clean the name (remove spaces + convert to title case)
    clean_name = student["name"].strip().title()
    
    # 2. Convert roll number to integer
    roll_no = int(student["roll"])
    
    # 3. Convert marks string into list of integers
    marks_list = []
    marks_split = student["marks_str"].split(", ")
    
    for mark in marks_split:
        marks_list.append(int(mark))
    
    # 4. Validate name (only alphabets in each word)
    is_valid = True
    for word in clean_name.split():
        if not word.isalpha():
            is_valid = False
    
    # Print validation result
    if is_valid:
        print(f"{clean_name} → ✓ Valid name")
    else:
        print(f"{clean_name} → ✗ Invalid name")
    
    # 5. Print formatted student profile
    print("================================")
    print(f"Student : {clean_name}")
    print(f"Roll No : {roll_no}")
    print(f"Marks   : {marks_list}")
    print("================================\n")
    
    # 6. Special condition for roll number 103
    if roll_no == 103:
        print("Special Output for Roll 103:")
        print(clean_name.upper())
        print(clean_name.lower())
        print()
```

```python

# ================================
# PART 2 — MARKS ANALYSIS
# ================================

# Given data
student_name = "Ayesha Sharma"
subjects = ["Math", "Physics", "CS", "English", "Chemistry"]
marks = [88, 72, 95, 60, 78]

print("\n--- Subject-wise Grades ---\n")

# Function to assign grade
def get_grade(score):
    if 90 <= score <= 100:
        return "A+"
    elif 80 <= score <= 89:
        return "A"
    elif 70 <= score <= 79:
        return "B"
    elif 60 <= score <= 69:
        return "C"
    else:
        return "F"

# 1. Print subjects with grades
for i in range(len(subjects)):
    grade = get_grade(marks[i])
    print(f"{subjects[i]} : {marks[i]} → {grade}")

# 2. Total and average
total_marks = sum(marks)
average_marks = round(total_marks / len(marks), 2)

print("\nTotal Marks:", total_marks)
print("Average Marks:", average_marks)

# 3. Highest and lowest subject
highest = max(marks)
lowest = min(marks)

print("Highest Scoring Subject:", subjects[marks.index(highest)], "-", highest)
print("Lowest Scoring Subject :", subjects[marks.index(lowest)], "-", lowest)


# ================================
# WHILE LOOP — ADD NEW SUBJECTS
# ================================

count_new = 0

while True:
    subject = input("\nEnter subject name (or type 'done' to stop): ")

    if subject.lower() == "done":
        break

    marks_input = input("Enter marks (0-100): ")

    # Check if input is numeric
    if not marks_input.isdigit():
        print("⚠ Invalid input! Please enter a number.")
        continue

    marks_value = int(marks_input)

    # Check valid range
    if marks_value < 0 or marks_value > 100:
        print("⚠ Marks should be between 0 and 100.")
        continue

    # Add valid data
    subjects.append(subject)
    marks.append(marks_value)
    count_new += 1

# Final output after loop
print("\nNew subjects added:", count_new)
updated_avg = round(sum(marks) / len(marks), 2)
print("Updated Average:", updated_avg)
```

```python
# ================================
# PART 3 — CLASS SUMMARY
# ================================

# # Class data
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

print("\n--- CLASS REPORT ---\n")

# Header
print("Name              | Average | Status")
print("----------------------------------------")

pass_count = 0
fail_count = 0
all_averages = []

topper_name = ""
topper_avg = 0

# Loop through data
for name, marks in class_data:
    # Calculate average
    avg = round(sum(marks) / len(marks), 2)
    all_averages.append(avg)

    # Decide pass/fail
    if avg >= 60:
        status = "Pass"
        pass_count += 1
    else:
        status = "Fail"
        fail_count += 1

    # Check topper
    if avg > topper_avg:
        topper_avg = avg
        topper_name = name

    # Print formatted row
    print(f"{name:<18} | {avg:7.2f} | {status}")


# Final summary
print("\nPassed:", pass_count)
print("Failed:", fail_count)
print("Topper:", topper_name, "-", topper_avg)

# Class average
class_avg = round(sum(all_averages) / len(all_averages), 2)
print("Class Average:", class_avg)
```

```python
# ================================
# PART 4 — STRING OPERATIONS
# ================================

# Given essay
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# Step 1: Strip spaces
clean_essay = essay.strip()
print("\n1. Clean Essay:\n", clean_essay)

# Step 2: Title Case
title_case = clean_essay.title()
print("\n2. Title Case:\n", title_case)

# Step 3: Count "python"
count_python = clean_essay.count("python")
print("\n3. 'python' count:", count_python)

# Step 4: Replace "python"
replaced_text = clean_essay.replace("python", "Python 🐍")
print("\n4. Replaced Text:\n", replaced_text)

# Step 5: Split into sentences
sentences = clean_essay.split(". ")
print("\n5. Sentences List:\n", sentences)

# Step 6: Numbered sentences
print("\n6. Numbered Sentences:")
for i, sentence in enumerate(sentences, start=1):
    # Add full stop if missing
    if not sentence.endswith("."):
        sentence += "."
    print(f"{i}. {sentence}")
```
